```python
import pandas as pd
```

```python
df = pd.read_csv('Quan Ao.csv')
```

```python
df.head()
```

```python
def create_rich_description(row):
    desc = f"{row['name']}"
    if row['category']: desc += f" thuộc danh mục {row['category']}"
    if row['color']: desc += f", màu {row['color'][0] if isinstance(row['color'], list) else row['color']}"
    if row['gender']: desc += f", dành cho {row['gender']}"
    if row['price']: desc += f", giá {row['price']:,}đ"
    return desc

df['rich_description'] = df.apply(create_rich_description, axis=1)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Khuyến nghị dùng model này
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-mpnet-base-v2')

# Tạo embeddings batch để tăng tốc
descriptions = df['rich_description'].tolist()
embeddings = model.encode(descriptions, batch_size=32, show_progress_bar=True)
df['vector_db'] = embeddings.tolist()
```

```python
df.to_csv("data_Quan_Ao_with_vectordb.csv", index=False)
```
